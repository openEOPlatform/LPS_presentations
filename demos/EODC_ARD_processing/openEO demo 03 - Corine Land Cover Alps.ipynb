```python
from eo_utils import openeoMap
center = [47, 12.84]
zoom = 12

eoMap = openeoMap(center,zoom)
eoMap.map
```

```python
from openeo.util import to_bbox_dict
bbox = eoMap.getBbox()
spatial_extent = to_bbox_dict(bbox) 
print(spatial_extent)
```

```python
collection      = 'corine_land_cover'
corine_data = conn.load_collection(collection, spatial_extent=spatial_extent, bands=["band_1"], temporal_extent = ["2000-01-01", "2019-04-30"] )
```

```python
corine_2000 = corine_data.filter_temporal(extent=["2000-01-01", "2001-01-01"])
corine_2018 = corine_data.filter_temporal(extent=["2018-01-01", "2019-01-01"])
```

```python
corine_merged = corine_2000.merge_cubes(corine_2018)
```

```python
corine_merged_saved = corine_merged.save_result("NetCDF")
```

```python
corine_merged_saved.flat_graph()
```

```python
job = corine_merged_saved.create_job(title = "CORINE_LAND_COVER")
print(f"Batch job with id {job.job_id} has status  {job.status()}")
```

```python
job.start_job()
print(f"Batch job with id {job.job_id} has status  {job.status()}")
```

```python
import xarray as xr
import numpy as np
import matplotlib.pyplot as plt
output_data3 = xr.open_dataset("./demo_data/out_2000_12_15T00_00_00_EU010M_E049N015T1.nc")
output_data3 = output_data3.where(output_data3!=output_data3.attrs["nodata"], np.nan)

output_data4 = xr.open_dataset("./demo_data/out_2018_12_15T00_00_00_EU010M_E049N015T1.nc")
output_data4 = output_data4.where(output_data4!=output_data4.attrs["nodata"], np.nan)
```

```python
import matplotlib.colors as colors
plt.figure(figsize=(36,6))

plt.subplot(1,3,1)
plt.title("2000-12-15")
plt.imshow(output_data3.band_1, cmap = "gray")   
plt.colorbar()

plt.subplot(1,3,2)
plt.title("2018-12-15")
plt.imshow(output_data4.band_1, cmap = "gray")   
plt.colorbar()

plt.subplot(1,3,3)
vmin = -35
vmax = 22
norm = colors.TwoSlopeNorm(vmin=vmin, vcenter=0, vmax=vmax)
plt.title("Difference")
plt.imshow(output_data3.band_1 - output_data4.band_1, cmap = "RdBu", vmin=vmin, vmax=vmax, norm=norm)   
plt.colorbar()
```

# Finding what changed

Now we analyze the image to find what really changed...

```python
diff = output_data3.band_1 - output_data4.band_1
diff = diff.to_numpy()
corine_2000 = output_data3.band_1.to_numpy()
corine_2018 = output_data4.band_1.to_numpy()
```

```python
diff[diff != 0] = 1
diff[diff == 0] = np.nan
```

```python
import matplotlib

matplotlib.rcParams['figure.figsize'] = [15, 12]


fig, ax = plt.subplots(2)
fig.patch.set_facecolor('white')
im = ax[0].imshow((corine_2000*diff))
ax[1].imshow((corine_2018*diff))

#cbar0 = fig.colorbar(im, cax=ax)
#cbar0.ax.set_title(" °C ", fontdict={'fontsize':20})
```

```python
(unique, counts) = np.unique(corine_2000*diff, return_counts=True)
frequencies = np.asarray((unique, counts)).T
print(frequencies.astype("int64"))

(unique, counts) = np.unique(corine_2018*diff, return_counts=True)
frequencies = np.asarray((unique, counts)).T
print(frequencies.astype("int64"))
```
